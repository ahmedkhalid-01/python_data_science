```python
import pandas as pd
import os 
import warnings
```

```python
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

input_file_path = "C:/Python/python_not_titanic/Data Sorting/"
output_file_path = "C:/Python/python_not_titanic/"

csv_file_list = os.listdir(input_file_path)

#print (csv_file_list)
```

```python
df = pd.DataFrame()

for csv_files in csv_file_list:
    
    #print(csv_files)

    if csv_files.endswith(".csv"):
        
        col_data = csv_files.rsplit('.', 2)[-2]
        #print (col_data)
        #col_data_final = " ".join(col_data[3:-1])
        #print (col_data_final)
        
        df1 = pd.read_csv(input_file_path+csv_files)
            
        df1.insert(4, "Tag Name", col_data)
       
        #print(df1)
        
        df = df.append(df1)
        
        # Tag_Name = [col_data_final]
        # df['Tag Name'] = Tag_Name
        
        #print(df)

df.sort_values(by='RecordTime', ascending=True) 
        
df.to_csv(output_file_path + "New Master File.csv")

print(df)
```

```python
df2 = pd.read_csv(output_file_path + "New Master File.csv") ##, index_col = "RecordTime", parse_dates = True)
pivoted = df2.pivot_table(index = "RecordTime", columns = 'Tag Name', values = ['ValueAsReal'])
pivoted.index = pd.to_datetime(pivoted.index)

pivoted = pivoted.sort_index() 
```

```python
pivoted = pivoted.ffill(axis = 0)
#print(pivoted)

pivoted = pivoted.backfill(axis = 0)
#print(pivoted)

pivoted.to_csv(output_file_path + 'New Master File Pivoted.csv')

# pivoted.describe()
```
