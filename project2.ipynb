```python
import pandas as pd
import os
import warnings
```

```python
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

input_file_path = "C:/Python/python_not_titanic/Data Sorting/"
output_file_path = "C:/Python/python_not_titanic/"

csv_file_list = os.listdir(input_file_path)

#print (csv_file_list)
```

```python
df = pd.DataFrame(columns = ['RecordTime'])

for csv_files in csv_file_list:
    
    #print(csv_files)

    if csv_files.endswith(".csv"):
        
        col_name = csv_files.split(".")
        #print (k)
        #col_name = k[2:-1]
        #print (col_name)
        col_name_final = " ".join(col_name[2:-1])
        #print (col_name_final)
        
        df1 = pd.read_csv(input_file_path+csv_files, usecols = ["RecordTime","ValueAsReal"])
        df1.rename(columns={"ValueAsReal": col_name_final}, inplace=True)
        
        df = df.merge(df1, on = "RecordTime", how = 'outer')
        
        #del df1['StateDesc']
        #del df1['QualityDesc']
        

df.to_csv(output_file_path + "Combined File.csv") 

print (df)
```
