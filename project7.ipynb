```python
import pandas as pd
import sqlite3
```

```python
sqliteConnection = sqlite3.connect('C:\Python\python_not_titanic\pivoted_database.db')
sql_query = pd.read_sql_query ("""SELECT * FROM Pivoted_database LIMIT 10000""", sqliteConnection)
df = pd.DataFrame(sql_query)
display(df)
data = df.drop(axis=0, index=[0,1])
```

```python
data['ValueAsReal.48'] = pd.to_numeric(data['ValueAsReal.48'])
movingavg = data['ValueAsReal.48'].rolling(window=7).mean()
display(movingavg)
```

```python
df['Moving_average'] = movingavg
display(df)
conn = sqlite3.connect('C:\Python\python_not_titanic\Pivoted_database_average.db')
c = conn.cursor()
df.to_sql('Pivoted_database_average',conn,)
```
