```python
import pandas as pd
import os
import warnings
```

```python
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

input_file_path = "C:/Python/python_not_titanic/Data Sorting/"
output_file_path = "C:/Python/python_not_titanic/"

excel_file_list = os.listdir(input_file_path)

#print (excel_file_list)
```

```python
df = pd.DataFrame()

for excel_files in excel_file_list:
        
    if excel_files.endswith(".csv"):
        
        #print (excel_files)
        
        df1 = pd.read_csv(input_file_path+excel_files)
        
        df = df.append(df1)
        
        #print(len(df))

# df = pd.concat([df1,df])
        
df.sort_values(by='RecordTime', ascending=True)  

df.to_csv(output_file_path + "Master File.csv")

df
```
